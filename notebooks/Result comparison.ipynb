```python
import os

import numpy as np
import pandas as pd
import scipy.stats

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
DATA_PATH = os.path.expanduser('~/Documents/datasets/mulan')
```

```python
df = pd.read_csv(os.path.join(DATA_PATH, 'results'))
df
```

```python
sns.barplot(x='dataset', y='precision', hue='algorithm', data=df)
plt.ylim((0, 1))
plt.xticks(rotation=60);
```

```python
sns.barplot(x='dataset', y='recall', hue='algorithm', data=df)
plt.xticks(rotation=90);
```
